```python
import matplotlib.pyplot as plt

def main():
    sales = [100, 400, 300, 600]
    slice_labels = ['1st Quarter', '2nd Quarter', '3rd Quarter', '4th Quarter']
    
    plt.pie(sales, labels=slice_labels)
    
    plt.title('Sales by Quarter')
    plt.show()
    
if __name__=='__main__':
    main()
```
